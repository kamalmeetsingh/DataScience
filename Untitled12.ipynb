```r
review<-read.delim2("dataset/RestReviews.tsv", quote="",stringsAsFactor=FALSE)
```

```r
review
```

```r
class(review$Review)
```

```r
install.packages("tm")
```

```r
library(tm)
```

```r
corpus<-VCorpus(VectorSource(review$Review))
```

```r
as.character(corpus[[1]])
```

```r
corpus<-tm_map(corpus, content_transformer(tolower))
corpus<-tm_map(corpus, removeNumbers)
corpus<-tm_map(corpus, removePunctuation)
```

```r
install.packages("SnowballC")
```

```r
library(SnowballC)
```

```r
corpus<-tm_map(corpus, removeWords, stopwords())
```

```r
corpus<-tm_map(corpus, stemDocument)
```

```r
corpus<-tm_map(corpus, stripWhitespace)
```

```r
dtm<-DocumentTermMatrix(corpus)
```

```r
dtm
```

```r
dtm<-removeSparseTerms(dtm,0.999)
```

```r
dtm
```

```r
dtm_matrix<-as.data.frame(as.matrix(dtm))
```

```r
dtm_matrix
```

```r
class(dtm_matrix)
```

```r
dtm_matrix$likes<-as.factor(review$Liked)
```

```r
library(e1071)
```

```r
set.seed(3)
id<-sample(2,nrow(dtm_matrix),prob=c(0.7,0.3),replace = T)
traindata<-dtm_matrix[id==1,]
testdata<-dtm_matrix[id==2,]
```

```r
review_naive<-naiveBayes(likes~.,data=traindata)
review_naive
```

```r
pred_naive<-predict(review_naive,testdata,type="class")

```

```r
library(caret)
```

```r
confusionMatrix(table(pred_naive,testdata$likes))
```
