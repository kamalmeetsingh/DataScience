```r
model<-apriori(Adult,parameter=list(support=0.2, confidence=0.5), 
               appearance = list(default="rhs",
                                 lhs=c("capital-loss=None","native-country=United-States")))
```

```r
summary(model)
```

```r
install.packages("arulesViz", dependencies=TRUE)
```

```r
library(arulesViz)
plot(model)
```
