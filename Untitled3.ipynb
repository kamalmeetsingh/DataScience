```r
game<-read.csv("dataset/video_game.csv")
```

```r
game
```

```r
gamenew<-game[,c(7,8,9,10)]
gamenew
```

```r
?set.seed
```

```r
set.seed(4)
inTraing<-sample(2,nrow(gamenew),prob=c(0.7,0.3),replace=T)
trainset<-gamenew[inTraing==1,]
testset<-gamenew[inTraing==2,]
linermodel<-lm(trainset$Other_players~.,data = trainset)
linermodel
```

```r
predict<-predict(linermodel,testset)
```

```r
predict
```

```r
testsubset<-testset[1:100,]
plot(testsubset$Other_players[1:100], type="l")
lines(predict[1:100],col="red")
```

```r
employee<-read.csv("dataset/Employee_Data.csv")
summary(employee)
```

```r
employee$salary<-ifelse(employee$Emp_Sal==">50K","High","Low")
employee$salary<-as.factor(employee$salary)
employee<-employee[,-15]
employee
```

```r
set.seed(4)
nrow(employee)
ids<-sample(2,nrow(employee),prob=c(0.7,0.3),replace = T)
```

```r
trainset<-employee[ids==1,]
testset<-employee[ids==2,]

model<-glm(salary~.,data = trainset,family="binomial")
predvalues<-predict(model,testset,type="response")
table(predvalues>0.4,testset$salary)
```
