```r
game_data<-read.csv("dataset/video_game.csv")
#check data
game_data
```

```r
game_player<-game_data[,c(7,8,9,10)]
game_player
```

```r
set.seed(4)
training_data<-sample(2,nrow(game_player),prob=c(0.7,0.3),replace = T)
training_data_set<-game_player[training_data==1,]
test_data_set<-game_player[training_data==2,]
```

```r
linear_regression<-lm(data = training_data_set)
linear_regression
```

```r
predict<-predict(linear_regression, test_data_set)
```
