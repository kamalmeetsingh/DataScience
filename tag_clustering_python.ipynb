```python
import pandas as pd
import csv

#def load_link_data():
data = {}
#r = pd.read_csv("./dataset/links.csv")
r = csv.reader(open("./dataset/links.csv", 'r'))
for row in r:
     #print(row)
    data[row[0]] = row[1].split(',')
    print(data)
```

```python
#tag_data = load_link_data()
url_tag_data = data

# print (url_tag_data)
all_tags = []
all_urls = []

for url,tags in url_tag_data.items():
    all_urls.append(url)
    all_tags.extend(tags)

all_tags = list(set(all_tags)) # list of all tags in the space
print(all_tags)
```

```python
import numpy as np
numerical_data = [] # create vectors for each item
for url,tags in url_tag_data.items():
    v = []
    for t in all_tags:
        if t in tags: #kindle
            v.append(1)
        else:
            v.append(0)
            
    numerical_data.append(tuple(v))
data = np.array(numerical_data)

print(np.shape(data))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=60, random_state=0).fit(data)
labels = kmeans.labels_
cluster_centres = kmeans.cluster_centers_
```

```python
# print out the clusters
clustered_urls = {}
clustered_tags = {}
i = 0
for url in all_urls:
    clustered_urls.setdefault(labels[i], []).append(url)
    clustered_tags.setdefault(labels[i], []).extend(url_tag_data[url])
    i += 1

for cluster_id,urls in clustered_urls.items():
    print (cluster_id)
    print (urls)

for cluster_id,tags in clustered_tags.items():
    print (cluster_id)
    print (list(set(tags)))
```
