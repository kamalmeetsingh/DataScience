```r
data("Affairs")
```

```r
Affairs
```

```r
Affairs$yesno<-ifelse(Affairs$affairs>0,"YES","NO")
Affairs$yesno<-as.factor(Affairs$yesno)
class(Affairs$yesno)
```

```r
Affairs<-Affairs[-1]
set.seed(3)
id<-sample(2,nrow(Affairs),prob=c(0.7,0.3),replace = T)
traindata<-Affairs[id==1,]
testdata<-Affairs[id==2,]
```

```r
library(rpart)
```

```r
affairtree<-rpart(yesno~.,data=traindata)
affairtree
```

```r
plot(affairtree, margin=0.1)
text(affairtree,pretty=T,cex=0.7)
```

```r
predtree<-predict(affairtree,testdata,type="class")
predtree
```

```r
install.packages('e1071', dependencies=TRUE)
```

```r
library(caret)
```

```r
confusionMatrix(table(predtree,testdata$yesno))
```

```r
library(randomForest)
```

```r
affairforest<-randomForest(yesno~.,data=traindata)
affairforest
```

```r
predforest<-predict(affairforest,testdata,type="class")
predforest
```

```r
confusionMatrix(table(predforest,testdata$yesno))
```
