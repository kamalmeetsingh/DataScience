```r
insurance<-read.xlsx("dataset/InsuranceData.xlsx",1)
insurance
```

```r
kmodel<-kmeans(insurance,5)
kmodel
```

```r
set.seed(2)
```

```r
kmax=15
```

```r
wss<-sapply(1:kmax,function(k){kmeans(insurance,k,nstart=10)$tot.withinss})
```

```r
plot(1:kmax,wss,pch=16,xlab="clusters",ylab="total withinss",frame=F)
```
