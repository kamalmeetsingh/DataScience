```r
ds.ts<-ts(ds$X1962.01, start = c(1962,1),end=c(1975,12),frequency = 12)
```

```r
ds.train<-ts(ds$X1962.01, start = c(1962,1),end=c(1974,12),frequency = 12)
ds.test<-ts(ds$X1962.01, start = c(1975,1),end=c(1975,12),frequency = 12)
```

```r
library(forecast)
```

```r
ds.arima<-auto.arima(ds.ts)
ds.prd_arima<-forecast(ds.arima,h=12)
plot(forecast(ds.arima,h=12),include=1)
```

```r
ds.HW<-HoltWinters(ds.ts)
ds.HW_pred<-predict(ds.HW,n.ahead = 12)
ts.plot(ds.HW_pred)
```
